```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv(r"F:\Ritesh\IMDB_Movies.csv")
orgdata = movies
```

```python
movies.head()
```

```python
###Cleaning the data###
#Columnwise null count##
movies.isnull().sum(axis=0).sort_values(ascending=False)
```

```python
#Row wise null count#
movies.isnull().sum(axis=1).sort_values(ascending=False)
```

```python
#For columnwise null percentages#
movies.isnull().sum(axis=0).sort_values(ascending=False)/len(movies
) * 100
```

```python
#Dropping unwanted columns#
movies = movies.drop(['color','director_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes',
'actor_3_facebook_likes','actor_2_name','cast_total_facebook_likes','actor_3_name','duration','facenumber_in_poster','content_rating','country','movie_imdb_link',
'aspect_ratio','plot_keywords'], axis=1)
```

```python
movies.head()
```

```python
#Dropping the rows#
round(movies.isnull().sum(axis=0).sort_values(ascending=False)/len(
movies)*100,2)
```

```python
movies=movies[movies['gross'].notnull()]
movies=movies[movies['budget'].notnull()]
round(movies.isnull().sum().sort_values(ascending=False)/len(movies)*100,2)
```

```python
#Dropping the rows#
(movies.isnull().sum(axis=1).sort_values(ascending=False)>5).sum()
movies=movies[movies.isnull().sum(axis=1).sort_values(ascending=False)<=5]
```

```python
#Filing the missing nan values#
round(movies.isnull().sum().sort_values(ascending=False)/len(movies)*100,2)
```

```python
#We can see that the Language column has some NaN values, we will replace that with English since its in the Max#
movies.groupby('language').language.count().sort_values(ascending=False)
```

```python
movies.language = movies.language.fillna('English')
```

```python
###MOVIES WITH HIGHEST PROFIT###
#changing the unit of the budget and gross columns from $ to million $#
movies['budget']=movies['budget']/1000000
movies['gross']=movies['gross']/1000000
```

```python
movies.head()
```

```python
#creating new column name profit#
movies['profit']=movies['gross']-movies['budget']
```

```python
movies.head()
```

```python
#sorting everything in descending order#
movies.sort_values(by='profit',ascending=False)
```

```python
#Top 10 movies that made more profit#
top10 = movies.sort_values(by='profit',ascending=False).head(10)
```

```python
top10.head()
```

```python
#Dropping the duplicates#
movies.drop_duplicates(keep='first',inplace=True)
```

```python
top10.head()
```

```python
#Again top10 movies made more profit#
top10 = movies.sort_values(by='profit',ascending=False).head(10)
```

```python
top10.head()
```

```python
###IMDB TOP 250###
IMDb_Top_250=movies[movies['num_voted_users']>25000].sort_values(by='imdb_score',ascending=False).head(250)
```

```python
IMDb_Top_250.head()
```

```python
#Sorting it in descending order#
IMDb_Top_250['Rank']=IMDb_Top_250['imdb_score'].rank(method='first',ascending=False)
```

```python
IMDb_Top_250.head()
```

```python
#Top 250 movies in English Language#
IMDb_Top_250[IMDb_Top_250['language']!='English']
```

```python
###BEST DIRECTORS###
top10director=movies.groupby('director_name').imdb_score.mean().sort_values(ascending=False).head(10)
```

```python
###POPULAR GENRES###
TempGenre=movies.genres.str.split('|',expand=True).iloc[:,0:2]
TempGenre.columns=['genre_1','genre_2']
TempGenre.genre_2.fillna(TempGenre.genre_1,inplace=True)
```

```python
TempGenre.head()
```

```python
###CRITIC FAVOURITE AND AUDIENCE FAVOURITE###
Meryl_Streep = movies[movies['actor_1_name']=='Meryl Streep']
Leo_Caprio = movies[movies['actor_1_name']=='Leonardo DiCaprio']
Brad_Pitt = movies[movies['actor_1_name']=='Brad Pitt']
combined=Meryl_Streep.append([Leo_Caprio,Brad_Pitt])
```

```python
combined.groupby('actor_1_name')['num_critic_for_reviews','num_user_for_reviews'].mean()
```

```python
#CALCULATED DACADES#
movies['decade']=movies['title_year'].apply(lambda x: (x//10)*10).astype(np.int64)
movies['decade']=movies['decade'].astype(str)+'s'
movies=movies.sort_values(['decade'])
```

```python
#CREATED DATAFRAME BY DECADE#
df_by_decade=movies.groupby('decade')
df_by_decade['num_voted_users'].sum()
df_by_decade=pd.DataFrame(df_by_decade['num_voted_users'].sum())
```

```python
#plotted the bar graph for number of voted users vs decade#
df_by_decade.plot.bar(figsize=(15,8),width=0.8,hatch="//",edgecolor
='k')
plt.xlabel("Decade")
plt.ylabel("Voted Users")
plt.title("Voted users over Decades")
plt.yscale('log')
plt.show()
```
